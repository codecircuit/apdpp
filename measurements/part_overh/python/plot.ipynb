```python
from ggplot import *
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv("../data-2016-03-08.csv")
df["time/kernels[s]"] = df.apply(lambda x: x[1] / x[0], axis=1)
df.head()
```

```python
p = ggplot(aes(x="kernels", y="time[s]"), data=df)  + geom_point(size=20) + geom_line()\
+ ylab("Total Time [s]") \
+ xlim(0, 200) + ylim(0, 3) + ggtitle("1000 Iterations, N = 4096")
p
```

```python
p = ggplot(aes(x="kernels", y="time[s]"), data=df)  + geom_point(size=1) + geom_line()\
+ ylab("Total Time [s]") \
+ xlim(0, 8500) + ylim(0, 80) + ggtitle("1000 Iterations, N = 4096")
p
```

```python
from scipy import stats
import matplotlib.pyplot as plt

x = df["kernels"].values.tolist() # std python list
y = df["time[s]"].values.tolist() # std python list
slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)

fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, np.array(x) * slope)
print "Gradient of linear regression =", slope, "s/kernel"
print "relative time per additional kernel =", slope / df.iloc[0,1] * 100, "%"
```

```python
p = ggplot(aes(x="kernels", y="time/kernels[s]"), data=df)  + geom_point(size=5) + geom_line()\
+ ylab("Time per Kernel [s]") \
+ xlim(0, 8500) + ylim(0, 0.2) + ggtitle("1000 Iterations, N = 4096")
p
```
